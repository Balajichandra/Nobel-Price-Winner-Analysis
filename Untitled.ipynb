```python
#calculating USA dominating
data["USA_nobel"] = data["birth_country"] == "United States of America"
data["decade"] = (np.floor(data["year"]/10)*10).astype(int)
USA_Winners = data.groupby('decade',as_index=False)["USA_nobel"].mean()
USA_Winners
```

```python
sns.lineplot(x = USA_Winners["decade"],y = USA_Winners["USA_nobel"])
```

```python
#calculating womens in nobel winners
data["female_winners"] = data["sex"] == "Female"
female_winners = data.groupby(['decade','category'],as_index=False)['female_winners'].mean()
sns.lineplot(x = 'decade',y = 'female_winners',hue = 'category',data=female_winners)
```

```python
data.nsmallest(10,'year')
```

```python
#no of scientist who won more than twice
data.groupby('full_name').filter(lambda group : len(group) >= 2)
```

```python
#at which year most of scientist got price
data['birth_date'] = pd.to_datetime(data['birth_date'])
#age
data['age'] = data['year'] - data['birth_date'].dt.year
sns.scatterplot(x='year',y='age',data = data)
```

```python
#all category age plot
#plt.rcParams['figure.figsize'] = [11, 7]
sns.lmplot(x='year', y='age', data=data, row='category' ,lowess=True, aspect=2, line_kws={'color' : 'black'})
```

```python
#Youngest nobel price winners
res = data.nsmallest(1,'age')
print(res['full_name'],"\n",res['birth_date'],"\n",res['year'])
```

```python
#Oldest nobel price winner
old = data.nlargest(1,'age')
print(old['full_name'],"\n",old['birth_date'],"\n",old['year'])
```
